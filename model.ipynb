```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=["id","diagnosis"])
y=df["diagnosis"]

X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=1)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler 
X_scaler=MinMaxScaler().fit(X_train)
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(y_train)
encoded_y_train=le.transform(y_train)
encoded_y_test=le.transform(y_test)

y_train_categorical=to_categorical(encoded_y_train)
y_test_categorical=to_categorical(encoded_y_test)
```

```python
y_train_categorical.shape
```

# Train the Model



```python
model=Sequential()
model.add(Dense(units=100,activation="relu",input_dim=30))
model.add(Dense(units=100,activation="relu"))
model.add(Dense(units=2,activation="softmax"))
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.summary()
```

```python
callback=[EarlyStopping(monitor="val_loss",patience=2)]
model.fit(
    X_train_scaled,
    y_train_categorical,
    callbacks=callback,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss,model_accuracy=model.evaluate(
    X_test_scaled,
    y_test_categorical,
    verbose=2
)

print(f"loss: {model_loss}, accuracy: {model_accuracy}")
```

```python
encoded_prediction=model.predict_classes(X_test_scaled [:5])
prediction_labels=le.inverse_transform(encoded_prediction)
```

```python
print(f"predictions: {prediction_labels}")
print(f"actual: {list(y_test[:5])}")
```
